```python
def extract_data(format_data):
    # thumbnail=format_data['thumbnail']
    extentions=format_data['ext']
    format=format_data['format']
    url=format_data['url']
    audio=format_data['audio_ext']
  
    return {
        'extention':extentions,
        'audio':audio,
        'format':format,
        'url':url,
        'thumbnail':thumbnail,
    }
for i in formats:
    print(extract_data(i))

# extract_data(formats[19])
```
